# 📊 The Simpsons — Regresión Logística Binaria

**Tarea:** Clasificación binaria con Regresión Logística  
**Variable objetivo:** `rating_alto` (1 si imdb_rating ≥ 7.5, 0 si no)  
**Predictores:** `viewers_millions`, `season`, `duration_min`, `number_in_season`  
**Splits:** 80/20 · 70/30 · 60/40  

---

## 0. Imports y configuración

```python
import sys, os
sys.path.insert(0, os.path.abspath("../.."))
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report,
    ConfusionMatrixDisplay, roc_curve
)
from sklearn.preprocessing import StandardScaler
from config.database import engine

os.makedirs("../../data/graficas", exist_ok=True)
GRAFICAS = "../../data/graficas"

plt.rcParams.update({"figure.figsize": (12, 5), "axes.grid": True, "grid.alpha": 0.3})
sns.set_theme(style="whitegrid", palette="muted")
print("✅ Imports OK")
```

## 1. Carga y preparación de datos

```python
df = pd.read_sql("""
    SELECT episode_id, viewers_millions, imdb_rating,
           duration_min, season, number_in_season
    FROM fact_ratings
    ORDER BY episode_id
""", engine)

# Variable objetivo binaria: 1 = rating alto (>= 7.5), 0 = rating bajo
UMBRAL = 7.5
df["rating_alto"] = (df["imdb_rating"] >= UMBRAL).astype(int)

print(f"Total registros: {len(df)}")
print(f"\nDistribución de clases (umbral = {UMBRAL}):")
vc = df["rating_alto"].value_counts().rename({0: 'Rating bajo (<7.5)', 1: 'Rating alto (>=7.5)'})
print(vc)
print(f"\nBalance: {df['rating_alto'].mean()*100:.1f}% episodios con rating alto")

df.head()
```

```python
# Visualización del balance de clases
fig, ax = plt.subplots(figsize=(6, 4))
colores = ['#E53935', '#43A047']
vc_plot = df["rating_alto"].value_counts()
ax.bar(['Rating bajo (0)', 'Rating alto (1)'], vc_plot.values, color=colores, alpha=0.85)
ax.set_title('Distribución de clases — Variable objetivo', fontweight='bold')
ax.set_ylabel('Cantidad de episodios')
for i, v in enumerate(vc_plot.values):
    ax.text(i, v + 5, str(v), ha='center', fontweight='bold')
plt.tight_layout()
plt.savefig(f"{GRAFICAS}/logit_balance_clases.png", dpi=150, bbox_inches='tight')
plt.show()
```

```python
FEATURES = ["viewers_millions", "season", "duration_min", "number_in_season"]
TARGET   = "rating_alto"

df_model = df[FEATURES + [TARGET]].dropna()
X = df_model[FEATURES]
y = df_model[TARGET]

scaler = StandardScaler()

print(f"Registros para modelado: {len(df_model)}")
print(f"Features: {FEATURES}")
print(f"Target  : {TARGET}")
```

## 2. Función de métricas

```python
def metricas_logistica(y_true, y_pred, y_prob, nombre_split):
    return {
        "Split":     nombre_split,
        "Accuracy":  round(accuracy_score(y_true, y_pred), 4),
        "Precision": round(precision_score(y_true, y_pred, zero_division=0), 4),
        "Recall":    round(recall_score(y_true, y_pred, zero_division=0), 4),
        "F1":        round(f1_score(y_true, y_pred, zero_division=0), 4),
        "ROC-AUC":   round(roc_auc_score(y_true, y_prob), 4),
    }

resultados = []
```

## 3. Split 1 — 80/20

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y)

X_train1_sc = scaler.fit_transform(X_train1)
X_test1_sc  = scaler.transform(X_test1)

logit1 = LogisticRegression(max_iter=1000, random_state=42)
logit1.fit(X_train1_sc, y_train1)
pred1 = logit1.predict(X_test1_sc)
prob1 = logit1.predict_proba(X_test1_sc)[:, 1]

resultados.append(metricas_logistica(y_test1, pred1, prob1, "80/20"))

print("=== Split 80/20 ===")
print(f"  Train: {len(X_train1)} | Test: {len(X_test1)}")
print(classification_report(y_test1, pred1, target_names=['Rating bajo', 'Rating alto']))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

ConfusionMatrixDisplay(confusion_matrix(y_test1, pred1),
    display_labels=['Bajo (0)', 'Alto (1)']).plot(ax=axes[0], colorbar=False)
axes[0].set_title('Matriz de Confusión — Split 80/20')

fpr, tpr, _ = roc_curve(y_test1, prob1)
auc = roc_auc_score(y_test1, prob1)
axes[1].plot(fpr, tpr, color='#2196F3', lw=2, label=f'ROC (AUC = {auc:.3f})')
axes[1].plot([0,1],[0,1],'k--', lw=1)
axes[1].set_xlabel('Tasa de Falsos Positivos'); axes[1].set_ylabel('Tasa de Verdaderos Positivos')
axes[1].set_title('Curva ROC — Split 80/20'); axes[1].legend()

plt.suptitle('Regresión Logística — Split 80/20', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig(f"{GRAFICAS}/logit_split1.png", dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Split 2 — 70/30

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(
    X, y, test_size=0.30, random_state=42, stratify=y)

X_train2_sc = scaler.fit_transform(X_train2)
X_test2_sc  = scaler.transform(X_test2)

logit2 = LogisticRegression(max_iter=1000, random_state=42)
logit2.fit(X_train2_sc, y_train2)
pred2 = logit2.predict(X_test2_sc)
prob2 = logit2.predict_proba(X_test2_sc)[:, 1]

resultados.append(metricas_logistica(y_test2, pred2, prob2, "70/30"))

print("=== Split 70/30 ===")
print(f"  Train: {len(X_train2)} | Test: {len(X_test2)}")
print(classification_report(y_test2, pred2, target_names=['Rating bajo', 'Rating alto']))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

ConfusionMatrixDisplay(confusion_matrix(y_test2, pred2),
    display_labels=['Bajo (0)', 'Alto (1)']).plot(ax=axes[0], colorbar=False)
axes[0].set_title('Matriz de Confusión — Split 70/30')

fpr, tpr, _ = roc_curve(y_test2, prob2)
auc = roc_auc_score(y_test2, prob2)
axes[1].plot(fpr, tpr, color='#FF9800', lw=2, label=f'ROC (AUC = {auc:.3f})')
axes[1].plot([0,1],[0,1],'k--', lw=1)
axes[1].set_xlabel('Tasa de Falsos Positivos'); axes[1].set_ylabel('Tasa de Verdaderos Positivos')
axes[1].set_title('Curva ROC — Split 70/30'); axes[1].legend()

plt.suptitle('Regresión Logística — Split 70/30', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig(f"{GRAFICAS}/logit_split2.png", dpi=150, bbox_inches='tight')
plt.show()
```

## 5. Split 3 — 60/40

```python
X_train3, X_test3, y_train3, y_test3 = train_test_split(
    X, y, test_size=0.40, random_state=42, stratify=y)

X_train3_sc = scaler.fit_transform(X_train3)
X_test3_sc  = scaler.transform(X_test3)

logit3 = LogisticRegression(max_iter=1000, random_state=42)
logit3.fit(X_train3_sc, y_train3)
pred3 = logit3.predict(X_test3_sc)
prob3 = logit3.predict_proba(X_test3_sc)[:, 1]

resultados.append(metricas_logistica(y_test3, pred3, prob3, "60/40"))

print("=== Split 60/40 ===")
print(f"  Train: {len(X_train3)} | Test: {len(X_test3)}")
print(classification_report(y_test3, pred3, target_names=['Rating bajo', 'Rating alto']))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

ConfusionMatrixDisplay(confusion_matrix(y_test3, pred3),
    display_labels=['Bajo (0)', 'Alto (1)']).plot(ax=axes[0], colorbar=False)
axes[0].set_title('Matriz de Confusión — Split 60/40')

fpr, tpr, _ = roc_curve(y_test3, prob3)
auc = roc_auc_score(y_test3, prob3)
axes[1].plot(fpr, tpr, color='#4CAF50', lw=2, label=f'ROC (AUC = {auc:.3f})')
axes[1].plot([0,1],[0,1],'k--', lw=1)
axes[1].set_xlabel('Tasa de Falsos Positivos'); axes[1].set_ylabel('Tasa de Verdaderos Positivos')
axes[1].set_title('Curva ROC — Split 60/40'); axes[1].legend()

plt.suptitle('Regresión Logística — Split 60/40', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig(f"{GRAFICAS}/logit_split3.png", dpi=150, bbox_inches='tight')
plt.show()
```

## 6. Interpretación de coeficientes

```python
coef_df = pd.DataFrame({
    "Variable": FEATURES,
    "Coeficiente": logit1.coef_[0],
    "Odds Ratio": np.exp(logit1.coef_[0])
}).sort_values("Coeficiente")

print("Coeficientes e Interpretación (Split 80/20):")
print(coef_df.round(4).to_string(index=False))
print(f"\nIntercepto: {logit1.intercept_[0]:.4f}")
print("\nInterpretación:")
print("  Odds Ratio > 1 → aumenta la probabilidad de rating alto")
print("  Odds Ratio < 1 → disminuye la probabilidad de rating alto")

fig, ax = plt.subplots(figsize=(9, 4))
colors = ['#E53935' if v < 0 else '#43A047' for v in coef_df["Coeficiente"]]
bars = ax.barh(coef_df["Variable"], coef_df["Coeficiente"], color=colors, alpha=0.85)
ax.axvline(0, color='black', linewidth=0.8, linestyle='--')
ax.set_xlabel("Coeficiente (escala estandarizada)")
ax.set_title("Coeficientes — Regresión Logística (Split 80/20)", fontweight='bold')
for i, v in enumerate(coef_df["Coeficiente"]):
    ax.text(v + (0.01 if v >= 0 else -0.01), i, f"{v:.4f}",
            va='center', ha='left' if v >= 0 else 'right', fontsize=9)
plt.tight_layout()
plt.savefig(f"{GRAFICAS}/logit_coeficientes.png", dpi=150, bbox_inches='tight')
plt.show()
```

## 7. Validación cruzada

```python
X_sc_full = scaler.fit_transform(X)
logit_cv = LogisticRegression(max_iter=1000, random_state=42)

cv_accuracy = cross_val_score(logit_cv, X_sc_full, y, cv=5, scoring='accuracy')
cv_f1       = cross_val_score(logit_cv, X_sc_full, y, cv=5, scoring='f1')
cv_auc      = cross_val_score(logit_cv, X_sc_full, y, cv=5, scoring='roc_auc')

print("Validación Cruzada (5-fold):")
print(f"  Accuracy : {cv_accuracy.mean():.4f} ± {cv_accuracy.std():.4f}")
print(f"  F1       : {cv_f1.mean():.4f} ± {cv_f1.std():.4f}")
print(f"  ROC-AUC  : {cv_auc.mean():.4f} ± {cv_auc.std():.4f}")
```

## 8. SUMMARY — Tabla comparativa de los 3 splits

```python
df_res = pd.DataFrame(resultados)

print("\n" + "="*65)
print("  SUMMARY — REGRESIÓN LOGÍSTICA BINARIA")
print("="*65)
print(df_res.to_string(index=False))
print("="*65)
print(f"\nMejor Accuracy : {df_res['Accuracy'].max():.4f} (Split {df_res.loc[df_res['Accuracy'].idxmax(), 'Split']})")
print(f"Mejor F1       : {df_res['F1'].max():.4f} (Split {df_res.loc[df_res['F1'].idxmax(), 'Split']})")
print(f"Mejor ROC-AUC  : {df_res['ROC-AUC'].max():.4f} (Split {df_res.loc[df_res['ROC-AUC'].idxmax(), 'Split']})")

# Gráfica comparativa
metricas_plot = ["Accuracy", "F1", "ROC-AUC"]
fig, axes = plt.subplots(1, 3, figsize=(14, 5))
colores = ['#2196F3', '#FF9800', '#4CAF50']

for ax, met, color in zip(axes, metricas_plot, colores):
    bars = ax.bar(df_res["Split"], df_res[met], color=color, alpha=0.85, width=0.4)
    ax.set_title(met, fontweight='bold')
    ax.set_xlabel("Split")
    ax.set_ylim(0, 1.15)
    for bar, val in zip(bars, df_res[met]):
        ax.text(bar.get_x() + bar.get_width()/2,
                bar.get_height() + 0.02,
                f"{val:.4f}", ha='center', fontsize=10)

plt.suptitle("Comparativa de Splits — Regresión Logística", fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig(f"{GRAFICAS}/logit_summary.png", dpi=150, bbox_inches='tight')
plt.show()

df_res
```
